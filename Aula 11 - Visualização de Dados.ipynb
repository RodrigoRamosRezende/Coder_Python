# Tipos de Graficos
### Grafico Linhas


```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 8, 6, 4, 2]

plt.plot(x, y)

plt.show()
```

### Gráfico de barras

```python
x = ['A', 'B', 'C', 'D', 'E']
y = [3, 7, 1, 8, 4]
plt.bar(x, y)
plt.show()
```

### Gráfico de pizza

```python
labels = ['Maçãs', 'Bananas', 'Laranjas', 'Morangos']
quantidades = [20, 25, 10, 30]
plt.pie(quantidades, labels=labels)
plt.show()
```

### Gráfico de dispersão

```python
x = [1, 2, 3, 4, 5]
y = [10, 8, 6, 4, 2]
plt.scatter(x, y)
plt.show()
```

### Histograma

```python
dados = [1, 3, 3, 4, 4, 4, 5, 5, 6, 7, 8, 9]
plt.hist(dados, bins=12)
plt.show()
```

### Gráfico de área

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.fill_between(x, y)
plt.show()
```

# Configurações

```python
x = [1, 2, 3, 4, 5]
y = [10, 8, 6, 4, 2]

plt.plot(x, y)

# Definindo o título e etiquetas dos eixos
plt.title("Gráfico de dispersão")
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")

plt.show()
```

```python
import matplotlib.pyplot as plt

# Dados para o gráfico de linha
x = [-1, 2, 3, 4, 5]
y = [-3, 5, 7, 9, 11]

# Configurações do gráfico
plt.plot(x, y, 
         color='red', 
         linestyle='-', 
         linewidth=2, 
         marker='o', 
         markersize=3, 
         markerfacecolor='white')
plt.title('Título do gráfico')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.xlim(0, 6)
plt.ylim(0, 12)
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Dados para o gráfico de linha
x = [-1, 2, 3, 4, 5]
y = [-3, 5, 7, 9, 11]

# Configurações do gráfico
plt.plot(x, y, 
         color='black', 
         linestyle='--', 
         linewidth=2, 
         marker='o', 
         markersize=8, 
         markerfacecolor='white')
plt.title('Título do gráfico')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.xlim(0, 6)
plt.ylim(0, 12)
# plt.grid(True)

plt.show()
```

# Subplot

```python
import numpy as np
# Dados para os gráficos de linha
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Criação de subplots de linhas
fig, axs = plt.subplots(3, 1, figsize=(30, 10))
axs[0].plot(x, y1)
axs[0].set_title('Gráfico de Seno')
axs[1].plot(x, y2)
axs[1].set_title('Gráfico de Cosseno')
axs[2].plot(x, y3)
axs[2].set_title('Gráfico de Tangente')

plt.show()
```
