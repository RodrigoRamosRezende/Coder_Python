```python
import pandas as pd

# 1. Lê o arquivo "best-selling game consoles.xlsx"
consoles_df = pd.read_excel("best-selling game consoles.xlsx", sheet_name='consoles')

# Conta quantos consoles há por companhia
contagem_consoles_por_companhia = consoles_df['Company'].value_counts().reset_index()
contagem_consoles_por_companhia.columns = ['Company', 'Count']

# 2. Ordena o resultado pela quantidade de consoles por companhia de forma decrescente
contagem_consoles_por_companhia = contagem_consoles_por_companhia.sort_values(by='Count', ascending=False)

# 3. Lê o arquivo "company.txt" e adiciona a coluna "Market Value" ao dataframe anterior
company_df = pd.read_csv("company.txt", delimiter=',')
consoles_df = pd.merge(consoles_df, company_df[['Company', 'Market Value']], on='Company', how='left')

# 4. Calcula o valor médio de Market Value por console
valor_medio_market_value_por_console = consoles_df['Market Value'].mean()

print("Quantidade de consoles por companhia:")
print(contagem_consoles_por_companhia)

print("\nValor médio de Market Value por console:")
print(valor_medio_market_value_por_console)
```
